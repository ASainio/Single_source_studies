```python
%matplotlib notebook
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
LC = pd.read_csv ('https://asas-sn.osu.edu/sky-patrol/coordinate/044d7a1e-1073-4f3e-88c8-b18178ae4f34/export.csv')
```

```python
#lets filter the bad data out using magic that I don't fully understand
data = LC[(LC['Filter']=='g') * (LC['flux_err'] < 0.4)]
#data.head()
```

```python
#lets make the time look bit better
time = data['HJD'] - 2450000
#time.head()
```

```python
#lets also define the flux for later editing
flux = data['flux(mJy)']
#flux.head()
```

```python
#Lets define flux error for later editing
flux_err =data['flux_err']
```

```python
#lets set the figure 
fig, ax = plt.subplots(1,1,figsize=(10,5))
#lets set main tittle and axel titles
ax.set_title('ASASSN-21qj')
ax.set_xlabel('Epoch [MJD]')
ax.set_ylabel('Flux [mJy]')
#lets plot or .scatter
ax.scatter(time, flux, linewidth = 0, marker = '.',color = 'k', alpha = 0.5)
#And lets slim the figure x and y
ax.set_xlim(8400,9500) 
ax.set_ylim(0,15)
```
