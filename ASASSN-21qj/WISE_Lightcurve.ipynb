```python
%matplotlib notebook
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#I had simply modified IRSA IPAC table to CSV.
LC = pd.read_csv ('Wise.Csv')
```

```python
#filtering out pad  stuff with quality flags and SNR
LC2 = LC[(LC['w2rchi2'] < 20) & (LC['cc_flags'] == '0000') & (LC['w2snr'] >= 25)]
```

```python
w2mag = LC2['w2mpro']
```

```python
time = LC2['mjd']
```

```python
#lets set the fixure 
fig, ax = plt.subplots(1,1,figsize=(10,5))
#lets set main tittle and axel titles
ax.set_title('ASASSN-21qj')
ax.set_xlabel('Epoch [MJD]')
ax.set_ylabel('WISE w2mpro')
#lets plot or .scatter
ax.plot(time, w2mag, linewidth = 0, marker = '.',color = 'k', alpha = 0.5)
#And lets slim the figure x and y
ax.set_xlim(56500,59500) 
ax.set_ylim(12.5,10.5)
```
