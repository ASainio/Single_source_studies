```python
%matplotlib notebook
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from astropy.io import ascii
```

```python
#one should be able to replace file with latest AAVSO csv without braking the code.
data = ascii.read('vstar.csv')
```

```python
#print(data)
```

```python
#lets define different filter
#Cousins I
i = data[(data['col5']=='I') * (data['col3'] < 0.5)] 
i_mjd = i['col1'] -2400000
i_mag = i['col2']
i_err = i['col3']
#Johnson V
v = data[(data['col5']=='V') * (data['col3'] < 0.5)]
v_mjd = v['col1'] -2400000
v_mag = v['col2']
v_err = v['col3']
#Johnson B
b = data[(data['col5']=='B') * (data['col3'] < 0.5)]
b_mjd = b['col1'] -2400000
b_mag = b['col2']
b_err = b['col3']
```

```python
#print(i_mjd)
```

```python
fig, ax = plt.subplots(1,1,figsize=(12,5))  

ax.set_title('ASASSN-21qj AAVSO Source')
ax.set_xlabel('Epoch [MJD]')
ax.set_ylabel('magnitude')

ax.plot (i_mjd, i_mag, linewidth = 0, marker = '.',color = 'red', alpha = 0.5, label= "Cousins I")  
ax.plot (b_mjd, b_mag, linewidth = 0, marker = '.',color = 'blue', alpha = 0.5, label = "Johnson B")  
ax.plot (v_mjd, v_mag, linewidth = 0, marker = '.',color = 'green', alpha = 0.9, label = "Johnson V")  

plt.ylim(19, 12.4)
plt.xlim(59450, 59580)
ax.legend()
```
